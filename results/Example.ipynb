```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 20})
plt.rcParams["figure.figsize"] = [12,9]

x = np.linspace(-3, 3, 1000)
y = x**2
plt.plot(x,y, label=r"$E(s) = s^2$")
plt.plot(x, [min(9, 1/abs(2*i)) for i in x], label=r"$p(s) = \frac{1}{|2s|}$")
plt.legend(loc=1)
plt.xlabel("s")
plt.ylabel("E")
plt.savefig("landscale.png")
plt.show()
```
